Question 1: What is the relaship between World Temperature Change and Wolrd GDP Growth and How has it affected it
====

```python
world_gdp_temperature = pd.merge(world_gdp, temperature_world,
                                 on = ["Country Name","Year"])
world_gdp_temperature
```

```python
average_gdp_growth = world_gdp_temperature["GDP Growth (annual %)"].mean()
average_temperature_growth = world_gdp_temperature["Temperature anomaly"].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(world_gdp_temperature["Year"],
         world_gdp_temperature["GDP Growth (annual %)"],
         label = "GDP Growth (annual %)")
plt.plot(world_gdp_temperature["Year"],
         world_gdp_temperature["Temperature anomaly"],
         label = "Temperature anomaly")
plt.axhline(y=average_gdp_growth, color='r', 
            linestyle='--', 
            label='Average GDP Growth')
plt.axhline(y=average_temperature_growth, color='g', 
            linestyle='--', 
            label='Average Temperature Anomally')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Global Temperature Anomaly and GDP Growth Over Time')
plt.legend()
plt.show()
```

```python
correlation = world_gdp_temperature[["Temperature anomaly", "GDP Growth (annual %)"]].corr()
correlation
```

The World correlation is -0.378 however this is not enough to explain the relation between temp and gdp 
So We are going to extract 20 largest Country in Tempearature and GDP Growth

```python
annual_gdp = annual_gdp[annual_gdp["Country Code"].isin(temperature_country["Country Code"])]
```

```python
average_country_gdp = annual_gdp.groupby("Country Code")["GDP Growth (annual %)"].mean()
filtered_gdp = annual_gdp[annual_gdp["Country Code"].isin(average_country_gdp.nlargest(20).index)]
filtered_temp = temperature_country[temperature_country["Country Code"].isin(average_country_gdp.nlargest(20).index)]
merged_gdp_temp = pd.merge(filtered_gdp, filtered_temp, on= ["Country Code", "Year"]) 
merged_gdp_temp
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data = merged_gdp_temp, x = "Year", y = "Temperature anomaly",
             label = "Temperature anomaly")
sns.lineplot(data = merged_gdp_temp, x = "Year", y = "GDP Growth (annual %)", 
             label = "GDP Growth (annual %)")
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
cor = merged_gdp_temp[["GDP Growth (annual %)", "Temperature anomaly"]].corr()
cor
```
