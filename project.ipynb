```python
import pandas as pd
import geopandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
annual_gdp = pd.read_csv("annual_gdp.csv")
annual_gdp
```

```python
annual_temperature_country = pd.read_csv("annual_temperature_anomalies.csv")
annual_temperature_country
```

```python
world_gdp = annual_gdp[annual_gdp["Country Name"] == "World"]
world_gdp
```

```python
change_co2_annual = pd.read_csv("change_co2_annual_pct.csv")
change_co2_annual
```

```python
world_temperature_growth = pd.read_csv("temperature-anomaly.csv")
world_temperature_growth
```

```python
world_temperature_growth = world_temperature_growth[world_temperature_growth["Entity"] == "Global"]
world_temperature_growth = world_temperature_growth[world_temperature_growth["Year"] > 1960]
world_temperature_growth
```

```python
world_temperature_growth = world_temperature_growth.melt(id_vars=["Entity", "Code", "Year"], 
                                                         var_name="Indicator", 
                                                         value_name= "value")
world_temperature_growth
```

```python
annual_gdp = annual_gdp[annual_gdp["Country Code"].isin(annual_temperature_country["Code"])]
annual_gdp = annual_gdp[["Country Name", "Country Code", "Indicator Name"] + [str(year) for year in range(1980, 2021)]]
annual_gdp
```

```python
annual_gdp = annual_gdp.melt(id_vars=["Country Name", "Country Code", "Indicator Name"], 
                             var_name="Year", 
                             value_name="GDP Growth")
annual_gdp["Year"] = annual_gdp["Year"].astype(int)
annual_gdp
```

```python
world_gdp = world_gdp[["Country Name", "Country Code", "Indicator Name"] + [str(year) for year in range(1961, 2021)]]
world_gdp
```

```python
world_gdp = world_gdp.melt(id_vars=["Country Name", "Country Code", "Indicator Name"],
                           var_name="Year",
                           value_name="GDP Growth")
world_gdp["Year"] = world_gdp["Year"].astype(int)
world_gdp
```

1. 기온 상승이 세계 GDP 성장률과 어떤 관계를 가지고, 이에 미친 영향은 무엇인가?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x= "Year", y= "GDP Growth", data= world_gdp)
plt.xlabel("Year")
plt.ylabel("GDP growth (annual %)")
plt.title("World growth GDP")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x= "Year", y= "value", data= world_temperature_growth, hue= "Indicator")
plt.xlabel("Year")
plt.ylabel("Global average temperature anomaly relative to 1961-1990")
plt.title("Global average temperature change")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x= "Year", y= "GDP Growth", data= annual_gdp, hue= "Country Name")
plt.xlabel("Year")
plt.ylabel("GDP growth (annual %)")
plt.title("Each Country growth GDP")
plt.show() 
```

```python
plt.figure(figsize=(12, 6))
plt.plot(world_temperature_growth['Year'], world_temperature_growth['value'], label='Temperature Anomaly')
plt.plot(world_gdp['Year'], world_gdp['GDP Growth'], label='GDP Growth')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Global Temperature Anomaly and GDP Growth Over Time')
plt.legend()
plt.show()
```

```python
data = pd.merge(world_gdp, world_temperature_growth, on=['Year'])
data
```

```python
correlation = data[['value', 'GDP Growth']].corr()
correlation
```

```python
world_temperature = annual_temperature_country[annual_temperature_country["Entity"] == "World"]
world_temperature = world_temperature[world_temperature["Year"] >= 1960]
world_temperature
```

```python
plt.figure(figsize=(12, 6))
plt.plot(world_temperature['Year'], 
         world_temperature['Temperature anomaly'], 
         label='Temperature Anomaly')
plt.plot(world_gdp['Year'], world_gdp['GDP Growth'], label='GDP Growth')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Global Temperature Anomaly and GDP Growth Over Time')
plt.legend()
plt.show()
```

```python
data = pd.merge(world_gdp, world_temperature, on=['Year'])
correlation = data[['Temperature anomaly', 'GDP Growth']].corr()
correlation
```

2. 기온 상승이 각 국가

```python
annual_country_indcustry = pd.read_csv("annual_country_industry.csv")
annual_country_indcustry
```

```python
annual_country_indcustry = annual_country_indcustry[["Country Name", "Country Code", "Indicator Name"] 
                                                    + [str(year) for year in range(1961, 2021)]]
annual_country_indcustry = annual_country_indcustry.melt(id_vars=["Country Name", "Country Code", "Indicator Name"], 
                             var_name="Year", 
                             value_name="Industry value")
annual_country_indcustry
```

```python
annual_country_indcustry = annual_country_indcustry[
    annual_country_indcustry["Country Code"].isin(annual_temperature_country["Code"])]
annual_country_indcustry = annual_country_indcustry[annual_country_indcustry["Country Code"]!= "World"]
annual_country_indcustry
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x= "Year", y= "Industry value", data= annual_country_indcustry, hue= "Country Name")
plt.xlabel("Year")
plt.ylabel("Industry value")
plt.title("Each Country growth GDP")
plt.show
```
